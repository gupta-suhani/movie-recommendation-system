```python
import ast
```

```python
#Extracting the genres from genres feature
def extract(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres']=movies['genres'].apply(extract)
```

```python
#Extracting the keywords
movies.iloc[0].keywords
```

```python
movies['keywords']=movies['keywords'].apply(extract)
```

```python
#Extracting the cast 
def extract(obj):
    L = []
    count=0
    for i in ast.literal_eval(obj):
        if(count==4):
            break
        L.append(i['name'])
        count=count+1
    return L
```

```python
movies['cast']=movies['cast'].apply(extract)
```

```python
#Extract the name of the director from crew column 
# movies.iloc[0].crew
def extract_dir(obj):
    L = []
    for i in ast.literal_eval(obj):
        if(i['job']=='Director'):
            L.append(i['name'])
            break
    return L
```

```python
movies['crew']=movies['crew'].apply(extract_dir)
```

```python
movies
```

```python
#function to convert string into list 
def conv(str):
    return str.split( )
```

```python
movies['overview']=movies['overview'].apply(conv)
```

```python
movies.head()
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies.head()
```

```python
movies_new=movies[['id','title','tags']]
```

```python
movies_new.head()
```

```python
#convert the list back to string 
movies_new['tags']=movies_new['tags'].apply(lambda x:" ".join(x))
```

```python
movies.iloc[0].tags
```

```python
#convert the string to lowercase
movies_new['tags']=movies_new['tags'].apply(lambda x:x.lower())
```

```python
movies_new.head()
```

```python
#function to convert the words into their root form
def stem(text):
    l=[]
    for i in text.split():
        l.append(ps.stem(i))

    return " ".join(l)
        
```

```python
movies_new['tags']=movies_new['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv= CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors =cv.fit_transform(movies_new['tags']).toarray()
```

```python
vectors
```

```python
#get features
arr=cv.get_feature_names_out()
for i in range(5000):
    print(arr[i])
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
#Now we will calculate the cosine similarity between the vectors 
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
def recommend(movie):
    movie_index=movies_new[movies_new['title']==movie].index[0]
    distance = similarity[movie_index]
    top_5=sorted(list(enumerate(distance)),reverse=True,key=(lambda x:x[1]))[1:6]

    for i in top_5:
        print(movies_new.iloc[i[0]].title)
    
    
```
